```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Read the image in, notice the format
img = plt.imread('Dogs.jpg')
# print(img)
# The shape is 3D because it is an array with rbg colors
# img.shape

# The imshow function with plot shows the image
# We use vmin and vmax as 0 and 255 because pixel's can be between 0-255 for each of the RGB values
# plt.imshow(img, vmin=0, vmax=255)
# plt.show()

# Standardize between 0-1 by dividing by 255
img = img / 255

# Now use vmin = 0 and vmax = 1
# plt.imshow(img, vmin=0, vmax=1)
# plt.show()

# Hold the shape of the image in a variable for when we want to translate it back to 3D array for plotting
img_shape = img.shape

# Reshape into a 2D array for feeding into our KMeans model
X = img.reshape(img_shape[0]*img_shape[1], img_shape[2])

# Reshape our data back using 3 shape args we held onto
# reversed_img = X.reshape(img_shape[0],img_shape[1],img_shape[2])

# Just a check, see if the image is the same
# plt.imshow(reversed_img, vmin=0,vmax=1)
# plt.show()
```

```python
# number of clusters in this project is the number of colors that we will find as the “center” of different clusters of colors
# Create a model with 4 clusters
model = KMeans(n_clusters=4)
```

```python
# random.seed is deprecated
# if you need to seed (e.g., to make computations reproducible for tests), create a new RNG
model = model.fit(X)
```

```python
# report cluster centers
print(model.cluster_centers_)
```

```python
# attribute labels_ lists closest cluster center for each of the points
print(model.labels_)
```

```python
# if we index the cluster center by the labels
# we will get back the cluster center assigned to each point
Y = model.cluster_centers_[model.labels_]
print(Y)
```

```python
# Reshape the array
Y = Y.reshape(img_shape)
# show image replicated
plt.imshow(Y, vmin=0, vmax=1)
plt.show()
```

```python
# Try different number of clusters
model = KMeans(n_clusters=11).fit(X)
# find replicated image
Y = model.cluster_centers_[model.labels_]
Y = Y.reshape(img_shape)
# Plot replicated image
plt.imshow(Y, vmin=0, vmax=1)
plt.title("{} Clusters".format(11))
plt.show()
```

```python
#We can get the colors used if we were interested
plt.imshow([model.cluster_centers_])
plt.show()
```
