```python
# y = 2x+3
# x = [1,2,3,4,5]
# y = [5,7,9,11,13]
# mean squared error (cost function) = 1/n * sum((y_k-y_predicted)**2 for k in (1 to n))
# mean squared error (cost function) = 1/n * sum((y_k-(mx_k + b))**2 for k in (1 to n))
# m is slope, b is y-intercept
# slope is change in y / change in x, slope at 2 point is 2 points difference
# slop at 1 point is derivative of point coordinate
# partial derivative: when function relies on >2 variable, set one as 0 and derivate the other
# do this for MSE, partial derivative gives slope, for step need LEARNING RATE
# from b1 to b2 = b1 - learning rate * slope
# gradient decsent is algorithm that finds best fit line for given training data set
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 100
    n = len(x)
    learning_rate = 0.09
    for i in range(0,iterations):
        y_predicted = m_curr * x + b_curr
        cost = 1/n * sum([val**2 for val in (y-y_predicted)])
        m_derivative = -(2/n) * sum(x*(y-y_predicted))
        b_derivative = -(2/n) * sum(y-y_predicted)
        m_curr = m_curr - learning_rate * m_derivative
        b_curr = b_curr - learning_rate * b_derivative
        print(f'm = {m_curr}, b = {b_curr}, cost = {cost} at iteration {i}')
```

```python
gradient_descent(x,y)
```
