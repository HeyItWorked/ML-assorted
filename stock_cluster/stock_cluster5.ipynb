```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('GroupedSP500.csv',index_col=0)
# mean value of numerical columns, central index key isn't relevant
df.drop(['CIK'],axis=1,inplace=True)
sub_df = df.groupby('Group').mean(numeric_only=True)
```

```python
sub_df.plot.bar()
plt.show()
```

```python
# plot without index and intercept
no_index_const = sub_df.drop(['Index','const'],axis=1)

no_index_const.plot.bar()
plt.show()
```

```python
# plot pie chart each sectors and groups associated within
for sector in df['GICS Sector'].unique():
    group = df[df['GICS Sector']==sector]['Group']
    group.value_counts().plot.pie(title=sector,legend=True)
    plt.show()
```
